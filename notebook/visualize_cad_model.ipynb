```python
import open3d as o3d
import numpy as np
```

```python
model_paths = ['../dataset/model/art_melon_without_handle.STL', '../dataset/model/dura_bright.STL', '../dataset/model/paprika_base.STL']
```

```python
models = []
pcds = []
for path in model_paths:
    print(f'||| loading {path}')
    model = o3d.io.read_triangle_mesh(path)
    model.compute_triangle_normals()
    pcd = model.sample_points_poisson_disk(number_of_points=30000)
    models.append(model)
    pcds.append(pcd)
```

```python
mesh_frame = o3d.geometry.TriangleMesh.create_coordinate_frame(size=0.4, origin=[0,0,0])
```

```python
o3d.visualization.draw_geometries([pcds[0], mesh_frame])
```

```python
import open3d as o3d
import numpy as np
```

```python
pcd = o3d.io.read_point_cloud('../dataset/orbbec/pointcloud/1769136634631267/RGBDPoints_1769136634631267.ply')
frame_mesh_big = o3d.geometry.TriangleMesh.create_coordinate_frame(size=1000.0, origin=[0,0,0])
o3d.visualization.draw_geometries([pcd, frame_mesh_big])
```

```python
# scale 1/1000 mm -> m
pcd.scale(0.001, np.array([0,0,0]))
frame_mesh = o3d.geometry.TriangleMesh.create_coordinate_frame(size=1.0, origin=[0,0,0])
o3d.visualization.draw_geometries([pcd, frame_mesh])
```
